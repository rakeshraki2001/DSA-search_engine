```python
s=[]
s.clear()
```

```python
import os
import re
```

```python
directory = 'Problemstatements'
total_docs=0
# iterate over files in that directory
for filename in os.listdir(directory):
    total_docs+=1
    if total_docs>100:
        break
    file = open("./Problemstatements/" +filename,  "r",encoding="utf8")
    doc=file.read()
    doc=doc.replace('\n', ' ').replace('\r', '')
    s.append(doc)
```

```python
import math
import os
import re
import nltk
from numpy import append
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
keys_set = set()
all_keys=[]
for doc in s:
    tdoc=re.sub('[^A-Za-z]+', ' ', doc)
    state_doc=tdoc.split("Example")
    split_doc=state_doc[0].lower().split(" ")
    stop_words = set(stopwords.words('english'))
    keys=[]
    for w in split_doc:
        if w not in stop_words:
            keys.append(w)
    keys.sort()
    all_keys.append(keys)

    for key in keys:
        keys_set.add(key)
```

```python
keys_set = sorted(keys_set)
```

```python
print(keys_set)
```

```python
print((all_keys))
```

```python
t_fs=[]
for doc_keys in all_keys:
    tdic= dict.fromkeys(keys_set,0)
    for ke in doc_keys:
        tdic[ke]+=(1/len(doc_keys))
    t_fs.append(tdic) 

print(t_fs)
        
```

```python
id_f=dict.fromkeys(keys_set,0)
```

```python
print(id_f)
```

```python
for ke in keys_set:
    cnt=0
    for dic in t_fs:
        if(dic[ke]!=0):
            cnt+=1
    id_f[ke]=math.log(len(t_fs)/cnt)
```

```python
print(id_f)
```

```python
search_string="Given a 1-indexed array of integers numbers that is already sorted in non-decreasing order, find two numbers such that they add up to a specific target number. Let these two numbers be numbers[index1] and numbers[index2] where 1 <= index1 < index2 <= numbers.length. Return the indices of the two numbers, index1 and index2, added by one as an integer array [index1, index2] of length 2. The tests are generated such that there is exactly one solution. You may not use the same element twice. Your solution must use only constant extra space."
```

```python
scores=[]
scores.clear()
for i in range(len(t_fs)):
    temp=[]
    temp.clear()
    for ke in keys_set:
        temp.append(t_fs[i][ke]*id_f[ke])
    scores.append(temp)
```

```python
print(scores)
```

```python
tdoc=re.sub('[^A-Za-z]+', ' ', search_string)
state_doc=tdoc.split("Example")
split_doc=state_doc[0].lower().split(" ")
stop_words = set(stopwords.words('english'))
ss_keys=[]
ss_keys.clear()
for w in split_doc:
    if w not in stop_words:
        ss_keys.append(w)
ss_keys= sorted(ss_keys)
```

```python
print((ss_keys))
```

```python
ss_tfs=dict.fromkeys(keys_set,0)
print(ss_tfs)
```

```python
for ke in ss_keys:
    if ke in ss_tfs:
        ss_tfs[ke]+=1/(len(ss_keys))
```

```python
print(ss_tfs)
```

```python
ss_score=[]
ss_score.clear()
for ke in keys_set:
    ss_score.append(ss_tfs[ke]*id_f[ke])

print(ss_score)
```

```python

projection_score={}

for i in range(len(scores)):
    sum=0
    magA=0
    magB=0
    for j in range(len(ss_score)):
        sum+=(ss_score[j]*scores[i][j])
        magA+=(scores[i][j]**2)
        magB+=(ss_score[j]**2)
    
    projection_score[i+1]= (sum)/(math.sqrt(magA*magB))

print(dict(sorted(projection_score.items(), key=lambda item: item[1])))
```
